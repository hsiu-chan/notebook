```python
numbers=[]
for i in range(1,6):
    numbers.append(input(f"Please input number {i}"))

print(f"The list is: {numbers}")

sum=0
odd=0

for number in numbers:
    number=int(number)
    sum+=number
    odd+=number%2

print(f"average = {sum/5}")
print(f"Odd_num = {odd}")
print(f"even_num = {5-odd}")




```
